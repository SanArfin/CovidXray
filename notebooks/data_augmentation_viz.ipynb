```python
# "Viral Pneumonia"
class_example = "Viral Pneumonia"

# number of example image
id_image = 1210


norm_path = os.path.join(base_path_out, "normalized_xrays", class_example)
augmented_path = os.path.join(base_path_out, "augmented", class_example)

# define filenames
norm_image_file_name = class_example + "-" + str(id_image) + ".png"

aug1_image_file_name = class_example + "-" + str(id_image) + "_ag1" +  ".png"
aug2_image_file_name = class_example + "-" + str(id_image) + "_ag2" +  ".png"
aug3_image_file_name = class_example + "-" + str(id_image) + "_ag3" +  ".png"
aug4_image_file_name = class_example + "-" + str(id_image) + "_ag4" +  ".png"
aug5_image_file_name = class_example + "-" + str(id_image) + "_ag5" +  ".png"
aug6_image_file_name = class_example + "-" + str(id_image) + "_ag6" +  ".png"
aug7_image_file_name = class_example + "-" + str(id_image) + "_ag7" +  ".png"


# read images
img_norm   = cv2.imread(os.path.join(norm_path ,norm_image_file_name), cv2.IMREAD_GRAYSCALE)

aug1_image = cv2.imread(os.path.join(augmented_path, aug1_image_file_name), cv2.IMREAD_GRAYSCALE)
aug2_image = cv2.imread(os.path.join(augmented_path, aug2_image_file_name), cv2.IMREAD_GRAYSCALE)
aug3_image = cv2.imread(os.path.join(augmented_path, aug3_image_file_name), cv2.IMREAD_GRAYSCALE)
aug4_image = cv2.imread(os.path.join(augmented_path, aug4_image_file_name), cv2.IMREAD_GRAYSCALE)
aug5_image = cv2.imread(os.path.join(augmented_path, aug5_image_file_name), cv2.IMREAD_GRAYSCALE)
aug6_image = cv2.imread(os.path.join(augmented_path, aug6_image_file_name), cv2.IMREAD_GRAYSCALE)
aug7_image = cv2.imread(os.path.join(augmented_path, aug7_image_file_name), cv2.IMREAD_GRAYSCALE)


# plot 1 normalized image and 7 augmented images
fig, axs = plt.subplots(2, 4, figsize=(9, 10))
fig.suptitle(class_example)


## first row of subplot

# plot 1: normalized image
axs[0, 0].imshow(img_norm, cmap='gray')
#axs[0, 0].set_title(norm_image_file_name)
axs[0, 0].axis('off')

# plot 2: augmented image
axs[0, 1].imshow(aug1_image, cmap='gray')
#axs[0, 1].set_title(aug1_image_file_name)
axs[0, 1].axis('off')

# plot 3: augmented image
axs[0, 2].imshow(aug2_image, cmap='gray')
#axs[0, 2].set_title(aug2_image_file_name)
axs[0, 2].axis('off')

# plot 4: augmented image
axs[0, 3].imshow(aug3_image, cmap='gray')
#axs[0, 3].set_title(aug3_image_file_name)
axs[0, 3].axis('off')


## second row of subplot

# plot 5: augmented image
axs[1, 0].imshow(aug4_image, cmap='gray')
#axs[1, 0].set_title(aug4_image_file_name)
axs[1, 0].axis('off')

# plot 6: augmented image
axs[1, 1].imshow(aug5_image, cmap='gray')
#axs[1, 1].set_title(aug5_image_file_name)
axs[1, 1].axis('off')

# plot 7: augmented image
axs[1, 2].imshow(aug6_image, cmap='gray')
#axs[1, 2].set_title(aug6_image_file_name)
axs[1, 2].axis('off')

# plot 8: augmented image
axs[1, 3].imshow(aug7_image, cmap='gray')
#axs[1, 3].set_title(aug7_image_file_name)
axs[1, 3].axis('off')
```
